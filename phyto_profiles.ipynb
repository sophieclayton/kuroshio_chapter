```python
% matplotlib inline

import numpy as np
import pandas as pd
from math import *
import matplotlib.pyplot as plt
import gsw as sw

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams.update({'font.family': 'serif'})
rcParams.update({'font.serif': 'Times New Roman'})
```

```python
# import the data file with all variables

in_file = '../data/data_CHL_fg.csv'
data = pd.read_csv(in_file, sep = ',')

print data.columns
```

```python
nstn = pd.unique(data.station)

pvar = [u'NO3 (uM)', u'Pro (cells/l)', u'Syn (cells/l)',
       u'Picoeuks (cells/l)', u'Diatoms (cells/l)',
       u'Dinoflagellates (cells/l)', u'Haptophytes (cells/l)']

fig, ax = plt.subplots(1, len(pvar), sharey=True, figsize = (10,4.5), dpi = 300)

for n in nstn:
    pp = 0
    sub = data[data['station']==n].reset_index(drop = 'T')
#     print sub['S'][0], n, sub['depth (m)'][0]
    if (sub['S'][0] > 33.4) & (sub['S'][0] < 34.2):        
        for p in pvar:
#             print p
            if p == 'NO3 (uM)':
                ax[pp-1].plot(sub[p], sub['depth (m)'])
            else:
                ax[pp-1].plot(np.log10(sub[p]), sub['depth (m)'])
            ax[pp-1].set_title(p)
            pp = pp+1
#             print pp
            
ax = plt.gca()
ax.set_ylim(0, 100)
ax.invert_yaxis()


plt.show()
```
