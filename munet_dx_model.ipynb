```python
% matplotlib inline

import numpy as np
import pandas as pd
from math import *
from matplotlib import rc

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams.update({'font.family': 'serif'})
rcParams.update({'font.serif': 'Times New Roman'})
import matplotlib.pyplot as plt
```

```python
# set up the parameters for the model

day_to_s = 60*60*24

munet = np.arange(0.1, 2.6, 0.1)/day_to_s
u = 10**np.arange(-1.5, 0.25, 0.1)

mv, uv = np.meshgrid(munet, u, sparse=False, indexing='ij')
```

```python
fig1 = plt.figure(1, figsize = (8,6), dpi = 300)
levels = np.log10([0.5,1, 2, 5, 10, 25, 50, 100, 500, 1000])

plt.rc('text', usetex=True)
cf = plt.contourf(u, munet*day_to_s, np.log10(uv/(2*mv)/1000), levels = levels, cmap = 'Greys')
plt.axhline(0.5, color='r')
plt.axhline(1.5, color='r')

plt.axvline(1, color='r')

cb = plt.colorbar()
cb.set_ticks(levels, update_ticks=True)
cb.ax.set_yticklabels(10**levels, fontsize = 16)
cb.set_label(r'$\Delta$x (km)', fontsize = 16)
plt.tick_params(labelsize =  16)
plt.ylabel(r'$\mu_{NET}$ (d^{-1})', fontsize = 16, rotation = 90)
plt.xlabel(r'u (m s^{-1})', fontsize = 16)
plt.title('$\Delta$x  - 50\% increase in biomass', fontsize = 16)
plt.show()

fig1. savefig('../figures/dx_munet_u_offset.png', dpi = 300)
```
